```python
from statsbombpy import sb
```

```python
competitions = sb.competitions()
```

```python
matches = sb.matches(competition_id=43,season_id=106)
matches.head()
```

```python
matches.shape
```

```python
matches.columns.tolist()
```

```python
matches[['match_date', 'home_team', 'away_team', 'home_score', 'away_score', 'competition_stage']]
```

```python
matches[matches['home_score']>3][['match_date', 'home_team', 'away_team', 'home_score', 'away_score', 'competition_stage']]
```

```python
matches['total_goals'] = matches['home_score'] + matches['away_score']
```

```python
matches.head()
```

```python
matches[['home_team','away_team','home_score','away_score','total_goals']].head(10)
```

```python
matches[['home_team','away_team','home_score','away_score','total_goals']].sort_values('total_goals', ascending = False)
```

```python
matches.groupby('home_team')['home_score'].sum()
```

```python
matches.groupby('home_team').agg(matches_played = ('match_id','count'),goals_scored =('home_score','sum'),goals_conceded = ('away_score','sum')).sort_values('goals_scored',ascending = False)
```

```python
team_goals = matches.groupby('home_team')['home_score'].sum().sort_values(ascending =False).head(10)
```

```python
team_goals
```

```python
plt.figure(figsize=(10,6))
plt.bar(team_goals.index,team_goals.values,color='steelblue')
plt.title('Top 10 teams by Goals Scored (2022 Fifa world Cup)')
plt.xlabel('Team')
plt.ylabel('Goals Scored')
plt.xticks(rotation = 45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(matches['home_score'],matches['away_score'], alpha = 0.6, color = 'Red', s=100)
plt.title("Team 1 vs Team 2 Goals (2022 worldcup)")
plt.xlabel("Team 1 Goals")
plt.ylabel("Team 2 Goals")
plt.tight_layout()
plt.show()
```
