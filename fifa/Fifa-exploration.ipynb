```python
matches = sb.matches(competition_id=43,season_id=106)
matches.head()
```

```python
matches.shape
```

```python
matches.columns.tolist()
```

```python
matches[['match_date', 'home_team', 'away_team', 'home_score', 'away_score', 'competition_stage']]
```

```python
matches[matches['home_score']>3][['match_date', 'home_team', 'away_team', 'home_score', 'away_score', 'competition_stage']]
```

```python
matches['total_goals'] = matches['home_score'] + matches['away_score']
```

```python
matches.head()
```

```python
matches[['home_team','away_team','home_score','away_score','total_goals']].head(10)
```

```python
matches[['home_team','away_team','home_score','away_score','total_goals']].sort_values('total_goals', ascending = False)
```

```python
matches.groupby('home_team')['home_score'].sum()
```

```python
matches.groupby('home_team').agg(matches_played = ('match_id','count'),goals_scored =('home_score','sum'),goals_conceded = ('away_score','sum')).sort_values('goals_scored',ascending = False)
```
