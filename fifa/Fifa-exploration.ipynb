```python
Shots[Shots['shot_outcome'] == 'Goal'][['player','shot_type','possession_team']]
```

```python
regular_shots = Shots[(Shots['period'] <= 4)]
```

```python
regular_shots.shape
```

```python
regular_shots['is_goal'] = regular_shots['shot_outcome'] == 'Goal'
```

```python
player_stats = regular_shots.groupby('player').agg(total_shots = ('id','count'),total_goals = ('is_goal','sum'),total_xG = ('shot_statsbomb_xg','sum')).sort_values('total_shots',ascending = False)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(player_stats['total_xG'],player_stats['total_goals'],alpha = 0.6,color = 'red')
for player in player_stats.index:
    plt.annotate(player,xy=(player_stats.loc[player,'total_xG'],player_stats.loc[player,'total_goals']),xytext=(player_stats.loc[player,'total_xG']+0.02,player_stats.loc[player,'total_goals']+.02))
plt.title("Player Stats (Argentina V/s Netherlands Quarter Final WC 2022)")
plt.xlabel("xG")
plt.ylabel("Total Goals")
plt.tight_layout()
plt.show() 
```

```python
match_list = matches['match_id'].tolist()
```

```python
all_events = []
for match_id in match_list:
    match_events = sb.events(match_id =match_id)
    all_events.append(match_events)
    
    
        
```

```python
len(all_events)
```

```python
all_events_df = pd.concat(all_events,ignore_index = True)
```

```python
all_events_df
```

```python
all_events_df[all_events_df['type'] == 'Shot']
```

```python
all_events_shots = all_events_df[all_events_df['type'] == 'Shot']
all_events_shots
```

```python
all_events_shots['is_goal'] = all_events_shots['shot_outcome'] =='Goal'
```

```python
all_events_shots.groupby('player').agg(total_shots = ('id','count'),total_xG =('shot_statsbomb_xg', 'sum'),total_goals = ('is_goal','sum')).sort_values('total_shots',ascending = False)
```

```python
player_tournament_stats = all_events_shots.groupby('player').agg(total_shots = ('id','count'),total_xG = ('shot_statsbomb_xg','sum'),total_goals = ('is_goal','sum')).reset_index()
```

```python
player_tournament_stats['OverPerformxG'] = player_tournament_stats['total_goals'] -player_tournament_stats['total_xG']
```
